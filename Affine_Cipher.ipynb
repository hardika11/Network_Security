```python
import math
import string
lower= string.ascii_lowercase
upper= string.ascii_uppercase
lowercase={}
for i in range(0,len(lower)):
    lowercase[lower[i]]=i

uppercase={}
for i in range(0,len(upper)):
    uppercase[upper[i]]=i

def conversion(pt):
    flag = []
    converted=[]
    a=list(pt)
    converted=[]
    for i in range(0,len(a)):
        flag.append(0)
    for i in range(0,len(a)):    
        if a[i].islower():
            c=lowercase[a[i]]
            converted.append(c)
            flag[i]=0
        else:
            c=uppercase[a[i]]
            converted.append(c)
            flag[i]=1
    return converted,flag

def char_con(conv,flag):
    cipher=""
    for i in range(0,len(flag)):
        if flag[i]==0:
            for c,n in lowercase.items():
                if conv[i]==n:
                    cipher+=c
        else:
            for c,n in uppercase.items():
                if conv[i]==n:
                    cipher+=c
    return cipher

def egcd(a, b): 
    x,y, u,v = 0,1, 1,0
    while a != 0: 
        q, r = b//a, b%a 
        m, n = x-u*q, y-v*q 
        b,a, x,y, u,v = a,r, u,v, m,n 
        gcd = b 
    return gcd, x, y 

def modinverse(a, m):
    gcd, x, y = egcd(a, m) 
    if gcd != 1: 
        return None
    else: 
        return x % m 
    
def encrypt(text, key): 
    output=[]
    text1=text.replace(' ','')
    con,f=conversion(text1)
    for t in range(0,len(con)):
        output.append(( key[0]*(con[t]) + key[1] ) % 26)
    cipher=char_con(output,f)
    return cipher
           
def decrypt(cipher, key): 
    decrypt=[]
    decon,f = conversion(cipher)
    for c in range(0,len(decon)):
        decrypt.append((modinverse(key[0], 26)*(decon[c] - key[1])) % 26)
    decrypt_text=char_con(decrypt,f)
    return decrypt_text

text = input("Enter plain text: ")
key1 = int(input("Enter key1: ")) 
key2 = int(input("Enter key2: "))
key=[key1,key2] 
if (math.gcd(key1,26)==1):
    encrypted = encrypt(text, key)
    print('Encrypted Text: {}'.format(encrypted))  
    print('Decrypted Text: {}'.format(decrypt(encrypted, key)))  
else:
    print("invalid key")
```
